```python
log_path = Path("pipeline_header.log")
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s %(levelname)s %(name)s %(message)s",
    handlers=[
        logging.FileHandler(log_path, encoding="utf-8"),
        logging.StreamHandler(),
    ],
)
logger = logging.getLogger("document_header_pipeline")

header_bigquery_table = "lab-test-project-1-305710.court_data_2024.document_header_data"
batch_size = 500
max_workers = 5

logger.info("Starting header extraction header pipeline for %s rows", len(legal_documents_df))
try:
    extract_missing_document_headers_with_bigquery_tracking(
        dataset=legal_documents_df,
        batch_size=batch_size,
        bigquery_table_id=header_bigquery_table,
        max_workers=3,
    )
except Exception:
    logger.exception("Pipeline failed")
    raise
```
