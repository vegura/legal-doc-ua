```python
import logging
from pathlib import Path

import pandas as pd

from document_pipeline_orchestrator import upload_missing_documents_with_bigquery_tracking
```

```python
data_file = "./edrsr_data_2024/documents.csv"
legal_documents_df = pd.read_csv(data_file, sep="\t")
```

```python
legal_documents_df.head()
```

```python
log_path = Path("pipeline.log")
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s %(levelname)s %(name)s %(message)s",
    handlers=[
        logging.FileHandler(log_path, encoding="utf-8"),
        logging.StreamHandler(),
    ],
)
logger = logging.getLogger("document_pipeline")

bucket_name = "court_data_2024"
bigquery_table_id = "lab-test-project-1-305710.court_data_2024.document_data"
batch_size = 500
max_workers = 5

logger.info("Starting pipeline for %s rows", len(legal_documents_df))
try:
    upload_missing_documents_with_bigquery_tracking(
        dataset=legal_documents_df,
        batch_size=batch_size,
        bucket_name=bucket_name,
        bigquery_table_id=bigquery_table_id,
        max_workers=max_workers,
        filename_id_column="doc_id",
        catalog_column="justice_kind",
    )
except Exception:
    logger.exception("Pipeline failed")
    raise
```
